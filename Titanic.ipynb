```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy
from array import *
plt.rc('xtick',labelsize=10) #set axis text size
plt.rc('ytick',labelsize=10)
df = pd.read_csv (r'C:\Users\user\Desktop\Dataset\Titanic.csv')   #read the csv file (put 'r' before the path string to address any special characters in the path, such as '\'). Don't forget to put the file name at the end of the path + ".csv"
a = df['CLASS'].values  #point
b = df['AGE'].values #Total
c = df['SEX'].values  #New
d = df['SURVIVE'].values  #Date
print("Total Number of Passengers",a.size)

#To find net survival ratio
count=0
for i in range(0,a.size):
    if d[i]==1:
        count=count+1
dead=a.size-count;
prob=(count/(count+dead))
print("Percentage of people survived is",prob*100,"%")    
labels = 'Survived', 'Dead'
sizes = [(prob*100),100-prob*100]
explode = (0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Net survived to died pie chart")
plt.show()

crew=0
first=0
second=0
third=0
for i in range(0,a.size):
    if(a[i]==0):
        crew=crew+1
    elif(a[i]==1):
        first=first+1
    elif(a[i]==2):
        second=second+1
    else:
        third=third+1
per_crew=(crew/a.size)*100
per_first=(first/a.size)*100
per_second=(second/a.size)*100
per_third=(third/a.size)*100
labels = 'Crew', 'FirstClass','SecondClass','ThirdClass'
sizes = [per_crew,per_first,per_second,per_third]
explode = (0,0,0,0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Class Percentage")
plt.show()

#User input Probability
print("Enter Class,Age,Sex to find survival probability:\n")
dead=0
total=0        
x=int(input("Enter Class:"))
y=int(input("Enter Age:"))
z=int(input("Enter Sex:"))
for i in range (0,a.size):
    if(a[i]==x and b[i]==y and c[i]==z):
        if(d[i]==0):
            dead=dead+1
        total=total+1
survived=total-dead        
print("Survived",total-dead)
print("Probability of survival",(total-dead)/total)

labels = 'Survival', 'Dead'
sizes = [survived, dead]
explode = (0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Required Pie")
plt.show()        
```
