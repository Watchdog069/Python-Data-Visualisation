```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
df = pd.read_csv (r'C:\Users\user\Desktop\Dataset\Salary.csv')   #read the csv file (put 'r' before the path string to address any special characters in the path, such as '\'). Don't forget to put the file name at the end of the path + ".csv"
x = df['YearsExperience'].values  #years
y = df['Salary'].values #salary
plt.rc('xtick',labelsize=8) #set axis text size
plt.rc('ytick',labelsize=8)
plt.plot(x,y,'go-', label='line 1', linewidth=1) #linewidth is width of the graph line
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Years of Experience vs Salary Graph')
plt.show()
n=x.size
sqr=lambda a:a*a
xsqr=0
xy=0
yi=0
xi=0
for i in range (0,n):
    xsqr=xsqr+sqr(x[i])
for i in range (0,n):
    yi=yi+y[i]
for i in range (0,n):
    xi=xi+x[i]
for i in range (0,n):
    xy=xy+x[i]*y[i]
A = np.array([[xsqr,xi], [xi,n]])
B = np.array([xy,yi])
C = np.linalg.solve(A, B)
print (C)   
a=C[0]
b=C[1]
print("enter input, years of experience:")
years=float(input())
salary=a*years+b
print("Salary expected=",salary)
```
